```python
import numpy as np
import pandas as pd
```

Below is the code to find out sample from the data we are going to use for doing our task.

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine_data = pd.read_csv(url, delimiter=';')
wine_data.head()
```

Below is code in which I have converted the value of quality attribute to 0 if the value present is smaller than or equal to 6 and 1 otherwise

```python
for i in range(len(wine_data)):
    if wine_data.loc[i, 'quality'] <= 6:
        wine_data.loc[i, 'quality'] = 0
    else:
        wine_data.loc[i, 'quality'] = 1
```

Here is how the dataset looks like after applying the above operation on quality attribute

```python
wine_data.head()
```

Below is the code where I have made a dataframe containing min and max values of each column

```python
wineDesc = wine_data.describe().iloc[[3, 7]]
wineDesc
```

Here I have just traversed through all the data points using for loops and iloc while skipping the values in quality attribute and applied the formula:
$$
    x_{s} = \frac{x_i - x_{min}}{x_{max} -x_{min}}
    \text{where }  x_{s}  \text{ is the scaled value.}
   $$
And after all i have printed out the final scaled data

```python
for i in range(len(wine_data)):
  for j in range(12):
    if(j == 11):
      continue
    wine_data.iloc[i, j] = (wine_data.iloc[i, j]-wineDesc.iloc[0, j]) / (wineDesc.iloc[1, j]-wineDesc.iloc[0, j])
wine_data.head()
```
